```python
from googleapiclient.discovery import build
# import pandas as pd
import pandas as pd

api_key = 'AIzaSyA53reIfoLIzgYoBxVAntEuFn9LwQJ7YUk'

def video_comments(video_id):

	# creating youtube resource object
  youtube = build('youtube', 'v3', developerKey=api_key)

	# retrieve youtube video results
  video_response=youtube.commentThreads().list(
	part='snippet',
	videoId=video_id,
  searchTerms="nikuba",
  textFormat="plainText").execute()

	# iterate video response
  while video_response:

		# extracting required info
		# from each result object
    for item in video_response['items']:

			# Extracting comments
      comment = item['snippet']['topLevelComment']
      author = comment["snippet"]["authorDisplayName"]
      text = comment["snippet"]["textDisplay"]
      datet = comment["snippet"]["publishedAt"]

      # append the comment to the lists
      comments.append(text)
      authors.append(author)
      datetime.append(datet)

		# Again repeat
    if 'nextPageToken' in video_response:
      video_response = youtube.commentThreads().list(
					part = 'snippet',
					videoId = video_id,
          searchTerms="nikuba",textFormat="plainText",pageToken=video_response['nextPageToken']).execute()
    else:
      break

# Enter video id
video_id = "R2AOTWvNImQ"

# put comments extracted in specific lists for each column
comments, authors, datetime = [], [], []

# Call function
video_comments(video_id)

df = pd.DataFrame(list(zip(datetime, authors, comments)),
               columns=['Datetime','Username', 'Text'])
```
